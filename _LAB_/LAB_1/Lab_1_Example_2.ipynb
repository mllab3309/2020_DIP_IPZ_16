```python
## Завантаження пакетів
import numpy as np
import matplotlib.pyplot as plt
import skimage.io as io
plt.rcParams['font.size'] = 18
```

```python
## Завантаження файлу зображення
filename = 'd:/_Doc_Vol_My_/_GIT_/2020/2020_Digita_Image_Processing/Test_Images/io_test_01.jpg'
test_im = io.imread(filename)
## Визначення стркутури та розміру зображення
print ('IMAGE SHAPE', test_im.shape, 'IMAGE SIZE', test_im.size)
rows_num = len(test_im)
print ('ROWS NUMBER', rows_num)
```

```python
## Вивід оррігінального зображення на екран
plt.imshow(test_im)
plt.show()
```

```python
## Визначення напівтоновоного  зображення
image_gray = np.zeros ( (rows_num , rows_num,3), dtype=np.uint8)
print ('GRAY SHAPE', image_gray.shape, 'GRAY SIZE', image_gray.size)

# Only BLUE channal
image_gray [:, :, 0] =  0
image_gray [:, :, 1] =  0
image_gray [:, :, 2] =  test_im [:, :, 2]

plt.figure(figsize=(5, 5)) ## ЗАДАЕМ РАЗМЕР ВЫВОДИМОЙ ФИГУРЫ
plt.title('IMAGE BLUE CHANNAL')
plt.imshow(image_gray)
plt.show()        

for i in  range (rows_num):
    for j in  range (rows_num):
        # Gray image
        image_gray [i, j, :] = 0.299*test_im [i, j, 0]+0.587*test_im [i, j, 1]+0.114*test_im [ i, j, 2]
## Вивід перетвореного зображення на екран
plt.figure(figsize=(5, 5)) ## ЗАДАЕМ РАЗМЕР ВЫВОДИМОЙ ФИГУРЫ
plt.title('IMAGE GRAY')
plt.imshow(image_gray)
plt.show()        
```

```python
## Зберігання перетвореного зображення до файлу
#### skimage.io.imsave(fname, arr, plugin=None, check_contrast=True, **plugin_args)
out_filename = 'd:/_Doc_Vol_My_/_GIT_/2020/2020_Digita_Image_Processing/Test_Images/io_test_02_gray.jpg'
io.imsave(out_filename,image_gray)
```

```python
## Зберігання тільки сірого до файлу в форматі BMP
image_gray_channal = np.zeros ( (rows_num , rows_num), dtype=np.uint8)
for i in  range (rows_num):
    for j in  range (rows_num):
        # Gray image
        image_gray_channal [i, j] = image_gray[i, j, 0]
```

```python
out_filename = 'd:/_Doc_Vol_My_/_GIT_/2020/2020_Digita_Image_Processing/Test_Images/io_test_02_gray_1channal.bmp'
io.imsave(out_filename,image_gray_channal)
```
