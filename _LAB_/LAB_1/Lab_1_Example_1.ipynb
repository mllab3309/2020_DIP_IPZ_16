```python
## Завантаження пакетів
import numpy as np
import matplotlib.pyplot as plt
import skimage.io as io
plt.rcParams['font.size'] = 18
```

```python
## Завантаження файлу зображення
filename = 'd:/_Doc_Vol_My_/_GIT_/2020/2020_Digita_Image_Processing/Test_Images/io_test_01.jpg'
test_im = io.imread(filename)
## Визначення стркутури та розміру зображення
print ('IMAGE SHAPE', test_im.shape, 'IMAGE SIZE', test_im.size)
rows_num = len(test_im)
print ('ROWS NUMBER', rows_num)
```

```python
## Вивід оррігінального зображення на екран
plt.imshow(test_im)
plt.show()
```

```python
## Визначення перетворенного зображення
image_crop_ = np.zeros ( (450, 450 , 3), dtype=np.uint8)
print ('IMAGE CROP SHAPE', image_crop_.shape, 'IMAGE CROP SIZE', image_crop_.size)
## Формування перетовреного зображення
for i in  range (450):
    for j in  range (450):
        image_crop_ [i, j, :] = test_im [ j , 450-i, :]
## Вивід перетвореного зображення на екран
plt.imshow(image_crop_)
plt.show()        
```

```python
## Зберігання перетвореного зображення до файлу
#### skimage.io.imsave(fname, arr, plugin=None, check_contrast=True, **plugin_args)
out_filename = 'd:/_Doc_Vol_My_/_GIT_/2020/2020_Digita_Image_Processing/Test_Images/io_test_01_crop.jpg'
io.imsave(out_filename,image_crop_)
```
